```python
from gradientai import Gradient


def main():
    gradient = Gradient()

    base_model = gradient.get_base_model(base_model_slug="nous-hermes2")

    new_model_adapter = base_model.create_model_adapter(
        name="Margimodel"
    )
    print(f"Created model adapter with id {new_model_adapter.id}")


    sample_query = "### Instruction:Margi Shah? \n\n ### Response:"
    print(f"Asking: {sample_query}")
    ## Before Finetuning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
    print(f"Generated(before fine tuning): {completion}")

    samples=[
        {"inputs":"### Instruction: Who is Margi Shah? \n\n### Response: Margi is a student from charusat University and currently planning for her masters in canada."},
        {"inputs":"### Instruction: Who is this person named Margi Shah? \n\n### Response: Margi Shah Like Data Science And AI And wants to make her future in this field"},
        {"inputs":"### Instruction: What do you know about Margi Shah? \n\n### Response: Margi Shah is a student creator who specializes in the field of Data Science and has linkedin profile https://www.linkedin.com/in/margi-shah-0b1813203/"},
        {"inputs":"### Instruction: Can you tell me about Margi Shah? \n\n### Response: Margi Shah is a youtuber,video creator,and a creator who loves Data Science And AI and LLM's"}
    ]

    ## Lets define parameters for finetuning
    num_epochs=3
    count=0
    while count<num_epochs:
      print(f"Fine tuning the model with iteration {count + 1}")
      new_model_adapter.fine_tune(samples=samples)
      count=count+1

    #after fine tuning
    completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
    print(f"Generated(after fine tuning): {completion}")
    new_model_adapter.delete()
    gradient.close()

if __name__ == "__main__":
    main()
```
